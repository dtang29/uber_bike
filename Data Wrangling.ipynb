```python
#dependencies
import pandas as pd
import numpy as np
import os
```

```python
#Set path
uber_bike_path = os.path.join('uber_bike_data.csv')

#Read in csv file as dataframe
uber_df = pd.read_csv(uber_bike_path)

#List of columns
uber_df.values
```

```python
len(uber_df)
```

```python
uber_df.head(5)
```

```python
#rename the columns and remove whitespaces
uber_df = uber_df.rename(columns={'Trip ID': 'trip_id', 'Duration': 'duration', 'Start Date': 'start_date', 
                                 'Start Station': 'start_station', 'Start Terminal': 'start_terminal', 
                                 'End Date': 'end_date', 'End Station': 'end_station', 'End Terminal': 'end_terminal',
                                 'Bike #': 'bike_num', 'Subscriber Type': 'subscriber_type', 'Zip Code': 'zip_code'})
```

```python
uber_df.head(10)
```

```python
uber_df.dtypes
```

```python
#convert start_date and end_date into date_time format
uber_df['start_date'] = pd.to_datetime(uber_df['start_date'])
uber_df['end_date'] = pd.to_datetime(uber_df['end_date'])
```

```python
#add new columns extracting start_hour and end_hour
uber_df['start_hour'] = uber_df['start_date'].dt.hour
uber_df['end_hour'] = uber_df['end_date'].dt.hour
```

```python
#add new columns extracting month
uber_df['start_month'] = uber_df['start_date'].dt.strftime('%b')
uber_df['end_month'] = uber_df['end_date'].dt.strftime('%b')
```

```python
#add a new column that converts duration to Hour:Minute format
uber_df['duration_hour_min'] = pd.to_datetime(uber_df['duration'], unit='s').dt.strftime('%H:%M')
```

# Google Places API - Reverse Geocoding 

```python
#Set path
places_api_results = os.path.join('places_api_results.csv')

#Read in csv file as dataframe
geocoded_addresses_df = pd.read_csv(places_api_results)

#List first 5
geocoded_addresses_df.head(5)

#grab required columns
geocoded_addresses_df = geocoded_addresses_df[['input_string', 'neighborhood']]
```

```python
#List first 5
geocoded_addresses_df.head(5)
```

```python
#Merge the geocoded addresses with uber_df to get start_neighborhood
uber_df_merge = uber_df.merge(geocoded_addresses_df, left_on='start_station', right_on='input_string', how='left')
```

```python
uber_df_merge.head(10)
```

```python
#rename the joined neighborhood to start_neighborhood
uber_df_merge = uber_df_merge.rename(columns={'neighborhood':'start_neighborhood'})
uber_df_merge.head(5)
```

```python
#Merge the geocoded addresses with uber_df to get end_neighborhood
uber_df_merge = uber_df_merge.merge(geocoded_addresses_df, left_on='end_station', right_on='input_string', how='left')
uber_df_merge = uber_df_merge.rename(columns={'neighborhood':'end_neighborhood'})
uber_df_merge.head(5)
```

```python
final_df = uber_df_merge[['trip_id', 'duration', 'duration_hour_min', 'start_date', 'start_month', 'start_hour', 'start_station', 'start_neighborhood', 'start_terminal',
                         'end_date', 'end_month', 'end_hour', 'end_station', 'end_neighborhood', 'end_terminal', 'bike_num', 'subscriber_type', 'zip_code', 
                         ]]

final_df.head(5)
```

```python
#Export final dataset to csv
# path = os.path.join('Cleaned Datasets', 'cleaned_registered_food_services.csv')
final_df.to_csv('cleaned_uber_data.csv',index=False)
```
